# **Basic Plots with Matplotlib**

### **Line Plot**

```python
#importing package
import matplotlib.pyplot as plt
```

```python
#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]
```

```python
#telling what to plot
plt.plot(year,pop)

#printing the plot
plt.show()
```

# **Customization**

### **Line Plot**

```python
#Axis Labels

#importing packages
import matplotlib.pyplot as plt

#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]

#plot
plt.plot(year,pop)

#labels
plt.xlabel('Year')
plt.ylabel('Population')

#show plot
plt.show()
```

```python
#Title

#importing packages
import matplotlib.pyplot as plt

#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]

#plot
plt.plot(year,pop)

#labels
plt.xlabel('Year')
plt.ylabel('Population')

#title
plt.title('World Population Projections')

#show plot
plt.show()
```

```python
#Ticks

#importing packages
import matplotlib.pyplot as plt

#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]

#plot
plt.plot(year,pop)

#labels
plt.xlabel('Year')
plt.ylabel('Population')

#title
plt.title('World Population Projections')

#ticks
plt.yticks([0,2,4,6,8,10])

#show plot
plt.show()
```

```python
#Ticks(2)

#importing packages
import matplotlib.pyplot as plt

#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]

#plot
plt.plot(year,pop)

#labels
plt.xlabel('Year')
plt.ylabel('Population')

#title
plt.title('World Population Projections')

#ticks
plt.yticks([0,2,4,6,8,10],
           ['0','2B','4B','6B','8B','10B'])

#show plot
plt.show()
```

```python
#Add more data

#importing packages
import matplotlib.pyplot as plt

#data
year=[1950,1970,1990,2010]
pop=[2.519,3.692,5.263,6.972]

#add more data
year= [1800, 1850, 1900] +year
pop = [1.0, 1.262, 1.650] +pop

#plot
plt.plot(year,pop)

#labels
plt.xlabel('Year')
plt.ylabel('Population')

#title
plt.title('World Population Projections')

#ticks
plt.yticks([0,2,4,6,8,10],
           ['0','2B','4B','6B','8B','10B'])

#show plot
plt.show()
```
