## **Dealing with missing values**

```python
#finding missing values
titanic.isnull().sum()
```

```python
#use drop.na method
titanic.dropna(subset=['deck'],axis=0, inplace=True)
```

```python
titanic.isnull().sum()
```

```python
#remove na from whole dataframe
titanic.dropna().isnull().sum()
```

**We have dropped all the missing values but this is not the only and best way to deal with it**

# **Replacing missing values with average**

```python
t1.isnull().sum()
```

```python
#finding average (mean)
mean = t1['age'].mean
mean
```

```python
#replacing nan with mean of the data (updating as well)
t1['age'] = t1['age'].replace(np.nan, mean)
```

```python
t1.isnull().sum()
```
