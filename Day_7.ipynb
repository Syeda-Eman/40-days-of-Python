# **3. Renaming columns**

```python
#renaming columns
df = pd.DataFrame({'A_col':[1,2,3], 'B_col': [4,5,6]})
df.rename(columns={'A_col': 'A', 'B_col':'B'}, inplace=True)
df
```

```python
#or you can rename like this
df.columns=['col_a','col_b']
df
```

```python
#to replace any specific character
df.columns=df.columns.str.replace('_',' ')
df
```

```python
#adding prefix to columns
df = df.add_prefix('column_')
df
```

```python
#adding suffix to columns
df = df.add_suffix('bc')
df
```

```python
df.columns = ['col_a', 'col_b']
df
```
