## üîπ Intermediate Python Concepts

### 1Ô∏è‚É£ List Comprehension (Advanced Way to Create Lists)

```python

# Traditional way to create squares

squares = []  # Empty list

for i in range(5):   # Loop from 0 to 4
    squares.append(i * i)  # Add square of number

print("Traditional:", squares)

# List Comprehension (Short & Powerful way)
# Syntax: [expression for item in iterable]

squares_comp = [i * i for i in range(5)]

print("List Comprehension:", squares_comp)
```

### 2Ô∏è‚É£ Advanced Functions

```python

# Function with default parameter

def greet(name="Guest"):  # Default value if no argument passed
    return "Hello " + name

print(greet())          # Uses default
print(greet("Huda"))    # Uses provided argument

# Function with multiple return values

def calculate(a, b):
    sum_value = a + b
    product = a * b
    return sum_value, product   # Returning multiple values

result = calculate(5, 3)

print("Sum:", result[0])
print("Product:", result[1])
```

### 3Ô∏è‚É£ Lambda Functions (Anonymous Functions)

```python

# Normal function

def add(a, b):
    return a + b

print(add(2, 3))

# Lambda version (short one-line function)

add_lambda = lambda a, b: a + b

print(add_lambda(2, 3))

# Using lambda with map()

numbers = [1, 2, 3, 4]

# map() applies function to every element
doubled = list(map(lambda x: x * 2, numbers))

print("Doubled:", doubled)
```

## üîπ Object-Oriented Programming (Advanced)

```python

# Creating a class

class Person:

    # Constructor method (runs automatically when object is created)
    def __init__(self, name, age):
        self.name = name  # Instance variable
        self.age = age

    # Method inside class
    def introduce(self):
        print("Hi, I am", self.name)

# Creating object (instance)
p1 = Person("Ali", 20)

p1.introduce()
```

### Inheritance (Very Important Concept)

```python

# Parent class
class Animal:

    def speak(self):
        print("Animal makes sound")

# Child class inherits from Animal
class Dog(Animal):

    def speak(self):  # Method overriding
        print("Dog barks")

d = Dog()

d.speak()  # Calls overridden method
```

## üîπ Decorators (Advanced Python Feature)

```python

# Decorators modify behavior of functions

def decorator_function(original_function):

    def wrapper():
        print("Something before function runs")
        original_function()
        print("Something after function runs")

    return wrapper

@decorator_function   # Applying decorator
def display():
    print("Hello from display function")

display()
```

## üîπ Generators (Memory Efficient Iteration)

```python

# Normal function returns all values at once

def normal_function():
    return [1, 2, 3]

print(normal_function())

# Generator function uses 'yield'

def generator_function():
    yield 1
    yield 2
    yield 3

gen = generator_function()

for value in gen:
    print("Generated:", value)
```

## üîπ Advanced Exception Handling

```python

try:
    num = int(input("Enter number: "))
    result = 10 / num
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input! Enter a number")

finally:
    print("Execution completed")
```

## üéØ Advanced Mini Project: Number Guessing Game

```python

import random  # Import random module

secret_number = random.randint(1, 10)  # Random number between 1 and 10

while True:

    guess = int(input("Guess the number (1-10): "))

    if guess == secret_number:
        print("Correct! üéâ")
        break  # Exit loop

    elif guess < secret_number:
        print("Too low!")

    else:
        print("Too high!")
```
