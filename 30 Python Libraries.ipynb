## üì¶ Altair
**About:** Declarative statistical visualization library.

```
import altair as alt
import pandas as pd
df = pd.DataFrame({'x':[1,2,3],'y':[3,2,1]})
chart = alt.Chart(df).mark_line().encode(x='x', y='y')
chart
```
