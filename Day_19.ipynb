# **Loops**

### **While loop**



*   Error starts at 50
*   Divide error by 4 on every run
*   Continue untill error is no longer > 1



```python
error = 50

while error > 1:
      error = error/4
      print(error)
```

What if we dont divide it?\
It will continue forever as the statement is always try making it an infinite loop.

```python
error = 50

while error > 1:
      print(error)
```

### **For loop**

```python
fam = [1.73, 1.68, 1.71, 1.89]
print(fam[0])
print(fam[1])
print(fam[2])
print(fam[3])
```

```python
for height in fam:
  print(height)
```

### **enumerate func**

```python
for index, height in enumerate(fam):
  print("index " +str(index) + ":" + str(height))
```

### **Loop over string**

```python
for c in "family":
  print(c.capitalize())
```
